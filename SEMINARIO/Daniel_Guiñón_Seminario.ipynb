El algoritmo sigue siendo muy lento, pero aún se puede mejorar el tiempo haciendo el procedimiento de una sola vez, creando un set con las posibilidades y haciendo un lookup:

```
%%time
def find_all():
    x = [i for i in range(1,10)]
    output_set = set()
    numbers = [str(i) for i in x]
    operators = ["+","-","*","/"]
    print("Calculando \n", end="")
    for values in permutations(numbers,5):
        for oper in permutations(operators):
            formula = "".join(o + v for o,v in zip([""]+list(oper),values))
            # Añadir todos los posibles outputs a un set
            output_set.add(eval(formula))
    return output_set

output = find_all()

for i in range(-100,100):
    if i in output:
        print(i)
    else:
        print(str(i) + 'No')
```

Finalmente se puede programar de forma más condensada, optimizando el tiempo y la eficiencia global del agloritmo, de la siguiente forma:

```
%%time
expression = [None] * 9
results = {eval(''.join(expression))
           for expression[::2] in permutations('123456789', 5)
           for expression[1::2] in permutations('+-*/')}
print(min(results), max(results))
print(set(range(-70, 79)) - results)
```

Esta implementación devuelve un diccionario en el que podemos resolver las preguntas objetivo de forma instantánea, aunque está limitada a los datos de éste problema concreto.
